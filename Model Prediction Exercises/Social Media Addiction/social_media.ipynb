```python
import pandas as pd
df = pd.read_csv('teen_phone_addiction_dataset.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
# Drop multiple columns from a pandas DataFrame
columns_to_drop = ['Name', 'Gender', 'Location', 'Phone_Usage_Purpose', 'School_Grade']
new_df = df.drop(columns=columns_to_drop)
# new_df = df.drop(df['Name, Gender, Location, Phone_Usage_Purpose'], axis='columns')
correlations = new_df.corr()['Addiction_Level'].sort_values(ascending=False)
print(correlations)
```

```python
columns_to_drop_1 = ['Name', 'Gender', 'Location', 'Phone_Usage_Purpose', 'School_Grade', 'Family_Communication', 'ID']
df = df.drop(columns=columns_to_drop_1)
df.head()
```

```python
# Add a new column 'Is_Addicted' using a lambda function
df['Is_Addicted'] = df['Addiction_Level'].apply(lambda x: True if x > 5 else False)
df.head()
```

```python
columns_to_drop_2 = ['Addiction_Level', 'Is_Addicted']
X = df.drop(columns=columns_to_drop_2)
y = df['Is_Addicted']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(LogisticRegression(), X, y,cv=3)
```

```python
cross_val_score(SVC(), X, y, cv=3)
```

```python
cross_val_score(RandomForestClassifier(), X, y, cv=3)
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
