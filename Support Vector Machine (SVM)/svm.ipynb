```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
# Convert the dataset to a pandas DataFrame
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

```python
df['target'] = iris.target
df.head()
```

```python
iris.target_names
```

```python
df['flower_name'] = df['target'].apply(lambda x: iris.target_names[x])
df.head()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
df0 = df[df.target == 0]
df1 = df[df.target == 1]
df2 = df[df.target == 2]
```

```python
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.scatter(df0['sepal length (cm)'], df0['sepal width (cm)'], color='red', label='setosa', marker='+')
plt.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'], color='blue', label='versicolor', marker='o')
plt.scatter(df2['sepal length (cm)'], df2['sepal width (cm)'], color='green', label='virginica', marker='*')
```

```python
plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')
plt.scatter(df0['petal length (cm)'], df0['petal width (cm)'], color='red', label='setosa', marker='+')
plt.scatter(df1['petal length (cm)'], df1['petal width (cm)'], color='blue', label='versicolor', marker='o')
plt.scatter(df2['petal length (cm)'], df2['petal width (cm)'], color='green', label='virginica', marker='*')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(['target', 'flower_name'], axis=1)
y = df['target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.svm import SVC
model = SVC(C=5)
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
